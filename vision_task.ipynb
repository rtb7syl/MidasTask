```python
#prepare dataset of images
#prepare train data of images from train pickle file


import pickle
import cv2
import os
import numpy as np


def create_dataset(imdata_path,labels_path,out_dir):

    imdata_file = open(imdata_path,'rb')
    imgs = pickle.load(imdata_file)
    imdata_file.close()

    labels_file = open(labels_path,'rb')
    labels = pickle.load(labels_file)
    labels_file.close()

    num_imgs = len(imgs)

    for i in range(num_imgs):

        img = imgs[i]
        img = np.asarray(img)
        print(img.shape)

        img = img.astype('float32')

        img = np.reshape(img,(28,28))
        print(img.shape)

        label = str(labels[i])

        out = os.path.join(out_dir,label,str(i)+'.png')
        print(out)
        cv2.imwrite(out,img)



imdata_path = 'Vision_task_dataset_public/train_image.pkl'
labels_path = 'Vision_task_dataset_public/train_label.pkl'
out_dir = 'data/train'
    
for label in ['0','2','3','6']:
    
    out = os.path.join(out_dir,label)
    if not os.path.exists(out):
        os.makedirs(out)

create_dataset(imdata_path,labels_path,out_dir)
```

```python
#create train val split
# create training data split 

import os,shutil
import numpy as np
from subprocess import check_output

def calculate_num_of_images_per_class(dir_):

    
    class_labels = os.listdir(dir_)
    num_of_images_of_each_class_dict = {}

    total_number_of_images = 0

    for class_label in class_labels:

        
        num_of_images_of_this_class = len(os.listdir(os.path.join(dir_,class_label)))
        total_number_of_images = total_number_of_images + num_of_images_of_this_class 


        #print("Number of images for the species",sub_folder,":",num_of_files)

        num_of_images_of_each_class_dict[class_label] = num_of_images_of_this_class
        
    print('total_number_of_images : ' + str(total_number_of_images))
    print('num_of_images_of_each_class_dict : ',num_of_images_of_each_class_dict)
        
    return (total_number_of_images,class_labels,num_of_images_of_each_class_dict)
    
    

def calculate_images_ratio_per_class(total_number_of_images,num_of_images_of_each_class_dict):
    
    #  returns a mapping of each class to the fraction of it's images present,
    # in the entire dataset
    
    for class_ in num_of_images_of_each_class_dict:
        
        num_of_images_of_each_class_dict[class_] = num_of_images_of_each_class_dict[class_] / total_number_of_images
    
    print('ratio_of_images_per_class_dict : ',num_of_images_of_each_class_dict)
    return num_of_images_of_each_class_dict


def validation_split(total_number_of_images,val_ratio,ratio_of_images_per_class_dict):
    
    # creates a val directory in ../data,with each subdir corresponding to each class
    
    train_path = "data/train"
    val_path = "data/val"
    
    os.mkdir(val_path)
    
    #num_of_val_images_per_class_dict = {}
    sum_val_imgs = 0
    
    for class_ in ratio_of_images_per_class_dict:
        
        path_to_img_files = train_path + '/' + class_
        
        path_to_val_img_files = val_path + '/' + class_ 
        
        os.mkdir(path_to_val_img_files)
 
        total_val_imgs = int(total_number_of_images*val_ratio)

        num_of_val_images_of_this_class = int(total_val_imgs * ratio_of_images_per_class_dict[class_])
        print(num_of_val_images_of_this_class)
    
        sum_val_imgs = sum_val_imgs + num_of_val_images_of_this_class
        
        

        
        
        
        #image_fnames_of_this_class_list = check_output(["ls", path_to_img_files]).decode("utf8").strip().split('\n')
        image_fnames_of_this_class_list = os.listdir(path_to_img_files)
        val_image_fnames_of_this_class_list = list(np.random.choice(image_fnames_of_this_class_list, size=num_of_val_images_of_this_class, replace=False))
        
        
        
        for image_fname in val_image_fnames_of_this_class_list:
            
            src = path_to_img_files + '/' + image_fname
            
            dst = path_to_val_img_files + '/' 
            
            shutil.move(src,dst)
            
        print('class ' +class_ + ' val imgs created....')
        
    print('FILES MOVED. VAL GENERATED...')
            
        
    
    

        
def main():
    
    dir_ = "data/train"
    val_split_ratio = 0.2
    
    total_number_of_images,class_labels,num_of_images_of_each_class_dict = calculate_num_of_images_per_class(dir_)
    
    ratio_of_images_per_class_dict = calculate_images_ratio_per_class(total_number_of_images,num_of_images_of_each_class_dict)
    
    
    validation_split(total_number_of_images,val_split_ratio,ratio_of_images_per_class_dict)
    
            
main()
```

```python
import os 

from keras.preprocessing.image import ImageDataGenerator

# data pre-processing and augmentation
#initialising the data generators


dir_ = 'data'
train_dir = os.path.join(dir_,"train")

val_dir = os.path.join(dir_,"val")
#test_dir = "../data/test_local"
    


train_datagen = ImageDataGenerator(
    
                    rescale=1./255,

                )


val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    
                                train_dir,
                                target_size=(28, 28),
                                color_mode='grayscale',
                                class_mode="categorical"
                            )


validation_generator = val_datagen.flow_from_directory(
    
                             val_dir,
                             target_size=(28, 28),
                             color_mode='grayscale',
                             class_mode="categorical")

print(train_generator.class_indices)
print(validation_generator.class_indices)
```

```python
import os
import pickle

import numpy as np

from keras import models
from keras.layers import Conv2D,BatchNormalization,Activation,Dense,Dropout,MaxPooling2D,Flatten
from keras import optimizers
from keras import losses
from keras import metrics

from keras.callbacks import ModelCheckpoint
from keras.models import Model




class CNNModel():

    # custom cnn model

    def __init__(self,ip_shape=(28,28)):

        self.ip_shape = ip_shape

    def create(self):

        print('[INFO].....Model started building......')
        model = models.Sequential()


        model.add(Conv2D(32, (3, 3), padding="same",input_shape=(self.ip_shape[0],self.ip_shape[1],1)))
        model.add(Activation("relu"))
        model.add(BatchNormalization())
        model.add(Conv2D(32, (3, 3), padding="same"))
        model.add(Activation("relu"))
        model.add(BatchNormalization())
        model.add(MaxPooling2D(pool_size=(2, 2)))
        model.add(Dropout(0.25))

        # second CONV => RELU => CONV => RELU => POOL layer set
        model.add(Conv2D(64, (3, 3), padding="same"))
        model.add(Activation("relu"))
        model.add(BatchNormalization())
        model.add(Conv2D(64, (3, 3), padding="same"))
        model.add(Activation("relu"))
        model.add(BatchNormalization())
        model.add(MaxPooling2D(pool_size=(2, 2)))
        model.add(Dropout(0.25))
        
        

        # first (and only) set of FC => RELU layers
        model.add(Flatten())
        model.add(Dense(128))
        model.add(Activation("relu"))
        model.add(BatchNormalization())
        model.add(Dropout(0.5))

        # softmax classifier
        model.add(Dense(4))
        model.add(Activation("softmax"))       


        return model


    def init(self,weights_path):

        model = self.create()
        #model.load_weights(weights_path)


        
        model.compile(loss='categorical_crossentropy',
                    
                    optimizer= optimizers.SGD(lr=0.001, momentum=0.9, decay=0.0, nesterov=True),
                    #optimizer=optimizers.Adam(lr=0.001),
                    
                    metrics=[metrics.categorical_accuracy,'acc'])

        

        print('[INFO].....Model initialised......')

        #print('[INFO].....Model summary......')
        print(model.summary())

        

        return model

    
    def train(self,train_generator,
                validation_generator,
                weights_path,
                checkpoint_path="cnn_model2_weights_checkpoint/weights-{epoch:02d}-{val_acc:.5f}.h5"
               ):


        model = self.init(weights_path=weights_path)

        print('[INFO].....Model started training......')

        checkpoint = ModelCheckpoint(checkpoint_path, monitor='val_acc', verbose=1, save_best_only=False, save_weights_only=True, mode='max', period=1)


        # Train the model 
        history = model.fit_generator(

            train_generator,
            steps_per_epoch = 200,
            epochs = 100,
            validation_data = validation_generator,
            validation_steps=50,
            callbacks = [checkpoint]
            
        )

        # dump the training history data in a pkl 
        with open('cnn_model2_weights_checkpoint/cnn_model1_history.pickle', 'wb') as handle:
            pickle.dump(history.history, handle)
            


        print('[INFO].....Model finished training......')

        print('[INFO].....Model history......')

        print(history.history)
        
        

    def predict(self,img,model):

        #predict class label either 0,1,2,3 and confidence score of the prediction 
        #of an input img (of shape (M,N,3)) after applying some transformations on it
        #the created model with loaded weights must be passed as an argument



        label = model.predict_classes(img)

        confidence = model.predict_proba(img)

        print('label,confidence',label,confidence)

        label = label[0]
        confidence = confidence[0][label]

        print('label,confidence',label,confidence)

        return label,confidence




cnn_model = CNNModel((28,28))
cnn_model.train(train_generator,validation_generator)
```

```python
# visualizing train/val loss & train/val acc vs #epochs

import matplotlib.pyplot as plt
import pickle



with open('cnn_model2_weights_checkpoint/cnn_model1_history.pickle', 'rb') as handle:
    
    history = pickle.load(handle)
    
    plt.plot(history['acc'])
    plt.plot(history['val_acc'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    # summarize history for loss
    plt.plot(history['loss'])
    plt.plot(history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()

```

```python
# Use model to predict test images


import numpy as np
import os
import pickle
import pandas as pd


def predict_test_imgs(test_data_path,weights_path,labels_dict):
    
    cnn_model = CNNModel((28,28))
    model = cnn_model.create()
    model.load_weights(weights_path)


    test_data_file = open(test_data_path,'rb')
    imgs = pickle.load(test_data_file)
    test_data_file.close()

    num_imgs = len(imgs)
    
    print(num_imgs)
    test_data_preds = []
    
    num_each_class = {'0':0,'2':0,'3':0,'6':0}
    

    for i in range(num_imgs):
        
        print(i)
        
        test_data_pred = {}
        
        img = imgs[i]
        #print(img)
        img = np.asarray(img)
        print(img.shape)

        img = img.astype('float32')/255

        img = np.reshape(img,(28,28,1))
        
        img = np.expand_dims(img, axis=0)

        print(img.shape)
        
        label,confidence = cnn_model.predict(img,model)
        label = labels_dict[str(label)]
        
        num_each_class[str(label)] = num_each_class[str(label)] + 1 
        
        test_data_pred['image_index'] = i
        test_data_pred['class'] = label
        
        print(test_data_pred)
        
        test_data_preds.append(test_data_pred)
        
    df = pd.DataFrame(test_data_preds)
    
    return df,num_each_class

pd.set_option('max_rows',3000)
test_data_path = 'Vision_task_dataset_public/test_image.pkl'      
weights_path = 'cnn_model2_weights_checkpoint/weights-55-0.89062.h5'

labels_dict = {'0':0,'1':2,'2':3,'3':6}
df,num_each_class = predict_test_imgs(test_data_path,weights_path,labels_dict)



print(num_each_class)
```
